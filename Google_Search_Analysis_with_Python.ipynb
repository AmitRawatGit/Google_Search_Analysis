```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import time
# install pytrends using the command : pip install pytrends
Trending_topics = TrendReq(hl='en-US', tz=360)
```

```python
kw_list=["Data Analytics"]
Trending_topics.build_payload(kw_list,cat=0, timeframe='today 12-m')
time.sleep(5)
```

```python
data = Trending_topics.interest_over_time()
data = data.sort_values(by="Data Analytics", ascending = False)
data = data.head(10)
print(data)
```

```python
kw_list = ["Data Analytics"]
Trending_topics.build_payload(kw_list, cat=0, timeframe='2024-01-01 2024-02-01', geo='', gprop='')
data = Trending_topics.interest_over_time()
data = data.sort_values(by="Data Analytics", ascending = False)
data = data.head(10)
print(data)
```

```python
data = Trending_topics.interest_by_region()
data = data.sort_values(by="Data Analytics",
                        ascending = False)
data = data.head(10)
print(data)
```

```python
data.reset_index().plot(x='geoName', y='Data Analytics',
                        figsize=(10,5), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```
